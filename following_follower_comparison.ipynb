```python
import pandas as pd #importing pandas
```

```python
#for followers file
df=pd.read_json("followers.json")
```

```python
df.head(10) #seeing the data
```

```python
print(df['media_list_data'].apply(type))
```

```python
df.shape
```

```python
#separating usernames from the list
df['value'] = df['string_list_data'].apply(lambda x: x[0]['value'])
```

```python
df.head(10)
```

```python
df.sort_values(by="value") #sorting on the username basis
```

```python
#creating new dataframe for followers
followers=df['value']
followers.head()
```

```python
followers.shape
```

```python
#for following file
d=pd.read_json("following.json")
```

```python
d.head() #seeing the data
```

```python
d.shape
```

```python
#checking the data type
print(d['relationships_following'].apply(type))
```

```python
#separating usernames from the dict
d['title'] = d['relationships_following'].apply(lambda x: x.get('title'))
```

```python
d.head()
```

```python
#creating new dataframe for followins and sorting it
following=d['title']
following.sort_values()
```

```python
#comparing following and followers
diff = followers[~followers.isin(following)]#checking the values in followers thats not in following
diff.shape
```

```python
print(diff) #printing the values in followers thats not in following
```

```python
diff1= following[~following.isin(followers)] #checking the values in following thats not in followers
diff1.shape
```

```python
print(diff1)
```

```python
diff1.head(101).sort_values()
```

```python
common = followers[followers.isin(following)] #common values in both
common.shape
```

```python
print(common)
```

```python
#exporting all files
diff.to_csv('follower_diff.csv', index=False)
diff1.to_csv('following_diff.csv',index=False)
common.to_csv('common.csv',index=False)
```
